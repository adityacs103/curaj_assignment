```python
import pandas as pd
yt = [138,136,152,127,151,130,119,153]
ft = [150.25,139.50,157.25,143.50,
      138,127.50,138.25,141.50]

df = pd.DataFrame({'Yt':yt,'Ft':ft})
df
```

```python
df['error(Yt-Ft)']=df['Yt']-df['Ft']
```

```python
df
```

```python
df['Absolute error(|Yt-Fy|)']=df['error(Yt-Ft)'].abs()
```

```python
df
```

```python
df['Sqared error((Yt-Ft)^2)']=df['Absolute error(|Yt-Fy|)']**2
```

```python
df
```

```python
df["percent error"]=df['Yt-Ft']/df['Yt']*100
```

```python
df
```

```python
df['Absolute percent error']=df['percent error'].abs()
```

```python
df
```

```python
print("mean error=",df['error(Yt-Ft)'].mean())
```

```python
print("mean Absolute error=",df['Absolute error(|Yt-Fy|)'].mean())
```

```python
print("mean squared error",df['Sqared error((Yt-Ft)^2)'].mean())
```

```python
print("mean percent error=",df['percent error'].mean())
```

```python
print("mean absolute error=",df['Absolute percent error'].mean())
```
