```python
import pandas as pd
df = pd.DataFrame({'Month':['Jan','Feb','Mar','Apr',
                            'May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
                   'Observed':[266.0,145.9,183.1,119.3,180.3,168.5,
                               231.8,224.5,192.8,122.9,336.5,185.9]})
df
```

```python
l1=list(df['Observed'].values)
```

```python
l1
```

```python
l2=[0,]
for i in range(len(l1)-2):
    l2.append((l1[i]+l1[i+1]+l1[i+2])/3)
l2.append(0)
```

```python
l2
```

```python
am_3=pd.DataFrame(l2,columns={'0':'3AM'})
```

```python
am_3.rename(columns={'0':'3AM'},inplace=True)
```

```python
am_3
```

```python
df1=df.join(am_3)
```

```python
df1
```

```python
l3=[0,0,]
```

```python
for i in range(len(l1)-4):
    l3.append((l1[i]+l1[i+1]+l1[i+2]+l1[i+3]+l1[i+4])/3)
l3.append(0)
l3.append(0)
```

```python
l3
```

```python
am_5=pd.DataFrame(l3)
```

```python
am_5
```
